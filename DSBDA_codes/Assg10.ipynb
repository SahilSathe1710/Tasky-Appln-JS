```python
import numpy as np
```

```python
Sorted = sorted(df['fare'])
```

```python
q1 = np.percentile(Sorted,25)
```

```python
q1
```

```python
q3 = np.percentile(Sorted,75)
```

```python
q3
```

```python
iqr = q3-q1
```

```python
iqr
```

```python
lb = q1 - (1.5*iqr)
lb
```

```python
ub = q3 + (1.5*iqr)
ub
```

```python
outliers = []
for i in Sorted:
    if(i<lb or i>ub):
        outliers.append(i)
print(outliers)
```
